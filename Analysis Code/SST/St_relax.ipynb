```python
import numpy as np
import pandas as pd
pd.options.display.float_format = '{:.6f}'.format
import matplotlib.pyplot as plt
#plt.style.use('classic')
#%pylab inline
#pylab.rcParams['figure.figsize'] = (10, 7)
```

```python
df1 = pd.read_csv("St_Average_q2.txt", sep='\s+', header=None,index_col=False,names=['time', 'st25', 'st50', 'st75','st100'])
df1.head()
```

```python
l = [0.1,1,10,100]

x=df1["time"].to_numpy()
y2=df1["st25"].to_numpy()
y3=df1["st50"].to_numpy()
y4=df1["st75"].to_numpy()
y5=df1["st100"].to_numpy()
st2=plt.plot(x,y2,"b-",label="25 fillers")
st3=plt.plot(x,y3,"g-",label="50 fillers")
st4=plt.plot(x,y4,"r-",label="75 fillers")
st5=plt.plot(x,y5,"c-",label="100 fillers")

plt.axhline(0.00, 0.0, 10000, c="black", ls='--')

plt.xlabel(r'Time after elongation ($\tau$)',fontsize=13)
plt.ylabel(r'stress $\sigma$($\tau$)',fontsize=13)
plt.xscale('log')
plt.gca().set_xticks(l)
plt.gca().set_xticklabels(l)
plt.rc("xtick", labelsize=13)
plt.rc("ytick", labelsize=13)
plt.legend(loc='upper right', fontsize=13)
plt.xlim(0.05,500)
plt.ylim(-0.2,1.0)
plt.savefig("stress_q2",dpi=1000)
plt.show()
```
