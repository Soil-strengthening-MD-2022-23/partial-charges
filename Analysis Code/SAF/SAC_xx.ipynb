```python
import numpy as np
import pandas as pd
pd.options.display.float_format = '{:.6f}'.format
import matplotlib.pyplot as plt
#plt.style.use('classic')
#%pylab inline
#pylab.rcParams['figure.figsize'] = (10, 7)
```

```python
df1 = pd.read_csv("Gt_Average_q3.txt", sep='\s+', header=None,index_col=False,names=['time', 'gt25', 'gt50', 'gt75', 'gt100'], skiprows=1)
```

```python
l = [1,10,100,1000]

x=df1["time"].to_numpy()

y2=df1["gt25"].to_numpy()
y3=df1["gt50"].to_numpy()
y4=df1["gt75"].to_numpy()
y5=df1["gt100"].to_numpy()

Gt2=plt.plot(x,y2,"b-",label="25 fillers", linewidth=2.0)
Gt3=plt.plot(x,y3,"g-",label="50 fillers", linewidth=2.0)
Gt4=plt.plot(x,y4,"r-",label="75 fillers", linewidth=2.0)
Gt5=plt.plot(x,y5,"c-",label="100 fillers", linewidth=2.0)

plt.axhline(0.00, 0.0, 10000, c="black", ls='--')

plt.xlabel(r'Correlation time ($\tau$)',fontsize=13)
plt.ylabel('Stress Auto Correlation G(t)',fontsize=13)
plt.xscale('log')
plt.gca().set_xticks(l)
plt.gca().set_xticklabels(l)
plt.rc("xtick", labelsize=13)
plt.rc("ytick", labelsize=13)
plt.legend(loc='upper right', fontsize=13)
plt.xlim(0.25,1000)
plt.ylim(-0.5,2.8)
plt.savefig("Gt_Eab_q3",dpi=1000)
plt.show()
```
